```python
#importing libraries
import numpy as np
import pandas as pd
```

**Data Pre-processing :**

```python
# Importing the dataset
df = pd.read_csv('/content/Iris.csv')
df
```

```python
df['Species'] = df['Species'].replace({'Iris-setosa':1, 'Iris-versicolor':2, 'Iris-virginica':3})
```

```python
df.head()
```

```python
df.tail()
```

```python
X = df.drop('Species', axis=1)
Y = df['Species']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

**Creating and Training KNN Model:**

```python
knn_clf = KNeighborsClassifier(n_neighbors=5, metric='euclidean', p=2)
knn_clf.fit(X_train, y_train)
```

**Predict Test Results:**

```python
y_pred = knn_clf.predict(X_test)
```

```python
pd.DataFrame({'Actual':y_test, 'Predicted':y_pred})
```

**Model Evaluation Metrics**

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score

score = accuracy_score(y_test, y_pred)
print(score)
```

```python
report = classification_report(y_test, y_pred)
print(report)
```

```python
m1 = confusion_matrix(y_test, y_pred)
print(m1)
```

```python
from sklearn.metrics import log_loss

# Get predicted probabilities for train and test sets
y_train_prob = knn_clf.predict_proba(X_train)
y_test_prob = knn_clf.predict_proba(X_test)

# Get predicted class labels for train and test sets using argmax
y_train_pred = np.argmax(y_train_prob, axis=1) +1 #add 1 to align with your class encoding
y_test_pred = np.argmax(y_test_prob, axis=1) +1 #add 1 to align with your class encoding


train_accuracy = accuracy_score(y_train, y_train_pred)
test_accuracy = accuracy_score(y_test, y_test_pred)

train_loss = log_loss(y_train, y_train_prob)
test_loss = log_loss(y_test, y_test_prob)

print("Train Accuracy:", train_accuracy)
print("Test Accuracy:", test_accuracy)
print("Train Loss:", train_loss)
print("Test Loss:", test_loss)
```
